```python
indicateurs = pd.read_csv(r"data\dvf2022.csv")
indicateurs = indicateurs.dropna(subset=["INSEE_COM"])
indicateurs
```

```python
dvf2021 = pd.read_csv(r"data\dvf2021.csv")
dvf2021 = (
dvf2021
    .dropna(subset=["INSEE_COM"])
    [["INSEE_COM", "Nb_mutations", "PrixMoyen", "Prixm2Moyen"]]
    .add_prefix('21_')
    .rename(columns={"21_INSEE_COM": "INSEE_COM"})
)
dvf2021
```

```python
indicateurs = (
    indicateurs
    .merge(dvf2021, on="INSEE_COM", how="left")
    )

# Remplissage des valeurs manquantes de l'année 2021 par celles de 2022
indicateurs["21_Nb_mutations"] = indicateurs["21_Nb_mutations"].fillna(indicateurs["Nb_mutations"])
indicateurs["21_PrixMoyen"] = indicateurs["21_PrixMoyen"].fillna(indicateurs["PrixMoyen"])
indicateurs["21_Prixm2Moyen"] = indicateurs["21_Prixm2Moyen"].fillna(indicateurs["Prixm2Moyen"])

# Indicateurs d'évolution
indicateurs["Evolution_21_22_Nb_mutations"] = (indicateurs["Nb_mutations"] - indicateurs["21_Nb_mutations"]) / indicateurs["21_Nb_mutations"]
indicateurs["Evolution_21_22_PrixMoyen"] = (indicateurs["PrixMoyen"] - indicateurs["21_PrixMoyen"]) / indicateurs["21_PrixMoyen"]
indicateurs["Evolution_21_22_Prixm2Moyen"] = (indicateurs["Prixm2Moyen"] - indicateurs["21_Prixm2Moyen"]) / indicateurs["21_Prixm2Moyen"]

# Centrer-réduire les indicateurs d'évolution
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
indicateurs[["Evolution_21_22_Nb_mutations", "Evolution_21_22_PrixMoyen", "Evolution_21_22_Prixm2Moyen"]] = scaler.fit_transform(indicateurs[["Evolution_21_22_Nb_mutations", "Evolution_21_22_PrixMoyen", "Evolution_21_22_Prixm2Moyen"]])
```

```python
indicateurs = indicateurs.drop(columns=["Unnamed: 0", "Annee"])
```
